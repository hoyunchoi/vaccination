```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

import sys
sys.path.append("../library")
sys.path.append("lib")

from defaultPlot import *
```

```python
ageStrat = pd.read_csv(os.path.join('data', 'meanField', 'deathRate', 'age.csv'), header=None)
ageStrat.columns=['vaccine', 'deathRate']
ageStrat = ageStrat.set_index('vaccine')
ageStrat = ageStrat.sort_index()
```

```python
rateStrat = pd.read_csv(os.path.join('data', 'meanField', 'deathRate', 'rate.csv'), header=None)
rateStrat.columns=['vaccine', 'deathRate']
rateStrat = rateStrat.set_index('vaccine')
rateStrat = rateStrat.sort_index()
```

```python
fig, ax = plt.subplots()

ax.plot(ageStrat.index, ageStrat['deathRate'], 'bo-')
ax.plot(rateStrat.index, rateStrat['deathRate'], 'ro-')

fig.show()

```
